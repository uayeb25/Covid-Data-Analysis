```python
import pandas as pd
import json
import requests
```

Read owid dataset to extract Europe information

```python
covid_owid = pd.read_csv("../data_source/owid-covid-data.csv")
covid_europe = covid_owid.loc[ covid_owid["continent"] == "Europe" , ["location"] ].drop_duplicates()
```

```python
key = "hSGVrADU7uKtN7ywOkzqSA==twRvVHdOPc2JMXxG" ##api.api-ninjas key
```

```python
countries = pd.DataFrame()
```

```python
for c_ in covid_europe["location"].array:
    api_url = 'https://api.api-ninjas.com/v1/country?name={0}'.format(c_)
    response = requests.get(api_url, headers={'X-Api-Key': key})
    if response.status_code == requests.codes.ok:
        country_extra_info = pd.DataFrame(response.json())
        country_extra_info["location"] = c_
        countries = pd.concat([ countries , country_extra_info ])
        print("{0} Completed!".format(c_))
    else:
        print("Error:", response.status_code, response.text)
```
